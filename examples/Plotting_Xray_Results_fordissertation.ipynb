```python
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import numpy as np
import h5py
from mpl_toolkits.axes_grid1.anchored_artists import AnchoredSizeBar
from matplotlib.font_manager import FontProperties
runs = ['RAFIKI_A','RAFIKI_B','RAFIKI_C',
        'RAFIKI_D','RAFIKI_E','RAFIKI_F',
        'RAFIKI_G','RAFIKI_H','RAFIKI_I', 'SIMBA']
names = ['RAFIKI-A','RAFIKI-B','RAFIKI-C',
        'RAFIKI-D','RAFIKI-E','RAFIKI-F',
        'RAFIKI-G','RAFIKI-H','RAFIKI-I', 'SIMBA']
# --- Load all data ---
maps = []
for run in runs:
    data_file = f'../testing/{run}_z01_ranges_dissertation_xraydat.hdf5'
    with h5py.File(data_file, 'r') as f:
        maps.append(np.array(f['image/image_dat'][:]))
bg = 1e-12  # small floor value for log scaling


maps = [np.array(m) for m in maps]

maps = [
    np.where((m == 0) | (~np.isfinite(m)), 1e-1, m)
    for m in maps
]
# --- Global color scale ---
vmin = 1e-1
vmax = max(np.max(m) for m in maps)
norm = colors.LogNorm(vmin=vmin, vmax=vmax)

# --- Figure and grid ---
fig, axes = plt.subplots(2, 5, figsize=(25, 10))

for i, ax in enumerate(axes.flat):
    im = ax.imshow(maps[i], norm=norm)

    ax.set_xticks([])
    ax.set_yticks([])
    ax.text(0.05, 0.95, names[i],
            transform=ax.transAxes,
            fontsize=20, 
            va='top', ha='left', color='white')
    ax.set_xlim(384-384/4,384+384/4)
    ax.set_ylim(384-384/4,384+384/4)
    # Add scalebar ONLY to first panel
    if i == 0:
        fontprops = FontProperties(size=30)
        scalebar = AnchoredSizeBar(ax.transData,
                                  27.19, '500 kpc', 'lower left',
                                  pad=0.1,
                                  color='black',
                                  frameon=True,
                                  size_vertical=3,
                                  fontproperties=fontprops)
        ax.add_artist(scalebar)

# --- Adjust layout to leave space for colorbar ---
plt.subplots_adjust(right=0.88)
plt.subplots_adjust(wspace=0, hspace=0)

# --- Add full-height colorbar on the right ---
cax = fig.add_axes([0.90, 0.10, 0.02, 0.80])  # [left, bottom, width, height]
cbar = fig.colorbar(im, cax=cax)
cbar.set_label('Counts', fontsize=40)
cbar.ax.tick_params(labelsize=30)  # smaller tick labels

plt.savefig('stackedxray_lowmass.png')
plt.show()
```
